```python
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
# Gender Distribution
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show
```

```python
#Grouping

df1=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
df1
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df, annot=True)
plt.title("Relation between Parent's Education And Student Scores")
plt.show()
```

```python
df=pd.read_csv("Expanded_data_with_more_features.csv")
df.head()
```

```python
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
plt.figure(figsize=(10,5))

ax=df['EthnicGroup'].value_counts().head(7).plot(kind="barh",title="GROUPS")
ax.bar_label(ax.containers[0])
plt.show()

```

```python
sns.boxplot(data=df,x="ReadingScore")
plt.title("Reading Score Plots")
plt.show()
```

```python
ax=df['ParentMaritalStatus'].value_counts().head(7).plot(kind='line',title="Marital Status")

plt.show
```

```python
ax=df['TransportMeans'].value_counts().head(7).plot(kind="bar",title="Means Of Transport")
plt.xlabel("Means")
plt.ylabel("Count")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
ax=df['MathScore'].value_counts().head().plot(kind="pie",title="Means Of Transport")
plt.show()
```
